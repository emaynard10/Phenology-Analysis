```python
def cleaning(dfs):
    for x in dfs:
        x.dropna()
        x.rename(columns = {'Total Plants Blooming':'Total Blooms'}, inplace=True)
    print(dfs)

dfs = [df1, df2, df3, df4]
cleaning(dfs)
```

```python
df4
```

```python
def plotting(dfs):
    for p in dfs:
        p.plot(x='Date', y=['Total Blooms'])

dfs = [df1, df2, df3, df4]
plotting(dfs)
```

```python
ax= df1.plot(x='Date', y=['Total Blooms'])
df2.plot(ax=ax, x='Date', y=['Total Blooms'])
df3.plot()
```

```python
# to set the plot size
plt.figure(figsize=(16, 8), dpi=150)
  
# using plot method to plot open prices.
# in plot method we set the label and color of the curve.
df1['Total Blooms'].plot(x='Date', label='Transect 1', color='orange')
df2['Total Blooms'].plot(x='Date', label='Transect 2', color= 'red')
df3['Total Blooms'].plot(x='Date', label='Transect 3', color= 'blue')
df4['Total Blooms'].plot(x='Date', label='Transect 4', color= 'green')  
# adding title to the plot
plt.title('Total Blooms Across All Transects')
  
# adding Label to the x-axis
plt.xlabel('Weeks')
plt.ylabel('Total Blooms')  
# adding legend to the curve
plt.legend()
```
